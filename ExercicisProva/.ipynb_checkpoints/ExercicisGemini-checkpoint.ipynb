```python
import pandas as pd
import re

unames = ['user_id', 'gender', 'age', 'occupation', 'zip']

users = pd.read_table('users.dat', sep='::', header=None, names=unames, engine='python') # NO ÉS un csv, i per alguna raó el . dat separa amb :: les dades, per això la comanda
rnames = ['user_id', 'movie_id', 'timestamp']
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('movies.dat', sep='::', header=None, names=mnames, engine='python', encoding='latin-1')

# ---------------------------------------------------------
# PAS 1: Extracció de l'any (Creació de nova columna)
# ---------------------------------------------------------
# Utilitzem extract() en lloc de findall() perquè volem el valor en una cel·la, no una llista.
# El regex (\d{4}) captura grups de 4 dígits.
# SINO seria així movie_years = movies['title'].str.findall(r'(\d{4})')
movies['year'] = movies['title'].str.extract(r'\((\d{4})\)', expand=False)

# ---------------------------------------------------------
# PAS 2: Neteja del Títol (Eliminació de l'any)
# ---------------------------------------------------------
# Utilitzem la teva lògica amb l'àncora $ per assegurar que només esborrem l'any del final.
# strip() addicional per seguretat.
#movie_titles = movies['title'].str.replace(r'\s*\(\d{4}\)\s*$', '', regex=True) # El dolar asegura que sigui final de frase
                                                                                # \s* Elimina els espais que no hagin de ser

movies['clean_title'] = movies['title'].str.replace(r'\s*\(\d{4}\)\s*$', '', regex=True).str.strip()

# ---------------------------------------------------------
# PAS 3: Tokenització (Llista de paraules)
# ---------------------------------------------------------
# Apliquem la lògica de re.findall sobre la columna neta i en minúscules.
# Utilitzem una lambda per aplicar-ho fila per fila.
movies['tokens'] = movies['clean_title'].str.lower().apply(lambda x: re.findall(r'\b\w+\b', x))

# ---------------------------------------------------------
# Resultat
print(movies)
```

```python
import pandas as pd
import queues
# CONVERSIÓ CRÍTICA A NUMÈRIC
# Assegurem que la columna 'year' siguin Ints per poder comparar amb el range()
movies['year'] = pd.to_numeric(movies['year'], errors='coerce').fillna(0).astype(int)

pq = queue.PriorityQueue()
from_year = int(movies['year'][movies['year'] > 0].min()) # Evitem el 0 si n'hi ha
to_year = int(movies['year'].max())

for year in range(from_year, to_year + 1):
    mask = (movies['year'] == year)
    num_movies = len(movies[mask])

    if num_movies > 0:
        # Simulació de Max-heap: invertir resultats
        pq.put((-num_movies, year))

print("--- Top 5 Anys amb més producció ---")
for i in range(5):
    if not pq.empty():
        count_neg, year = pq.get()
        print(f"#{i+1}: Any {year} amb {-count_neg} pel·lícules")
    else:
        break
```

'Regular expressions (regex) are patterns to match, search, and manipulate text. Below is a compact “what it does + example” summary of the core pieces you’ll use 95% of the time.  \n    ## Literal characters\n\n    - `abc` – Matches the exact substring “abc”.  \n      - Example: Pattern `cat` matches “cat” in “my cat is here”.[1][2]\n\n    ## Character classes\n\n    - `[abc]` – Any one of `a`, `b`, or `c`.  \n      - Example: `[cb]at` matches “cat” and “bat”.[3]\n    - `[^abc]` – Any character except `a`, `b`, or `c`.  \n      - Example: `[^0-9]` matches any non-digit.[3]\n    - `[a-z]` – Any lowercase letter from a to z (ranges).  \n      - Example: `[A-Z][a-z]+` matches “Hello”.[4][3]\n    - Shorthands:  \n      - `\\d` digit, `\\D` non-digit.  \n      - `\\w` word char (letters, digits, `_`), `\\W` non-word.  \n      - `\\s` whitespace, `\\S` non-whitespace.  \n      - Example: `\\d\\d/\\d\\d/\\d{4}` matches dates like “12/03/2025”.[5][4][3]\n\n    ## Quantifiers (how many times)\n\n    - `a*` – 0 or more `a`.  \n      - Example: `ba*` matches “b”, “ba”, “baaa”.[6][7][3]\n    - `a+` – 1 or more `a`.  \n      - Example: `ba+` matches “ba”, “baaa” but not “b”.[7][6][3]\n    - `a?` – 0 or 1 `a` (optional).  \n      - Example: `colou?r` matches “color” and “colour”.[6][7]\n    - `a{3}` – Exactly 3 `a`.  \n      - Example: `a{3}` matches “aaa” in “baaaac”.[6][3]\n    - `a{2,4}` – Between 2 and 4 `a`.  \n      - Example: `a{2,4}` matches “aa”, “aaa” or “aaaa”.[3][6]\n\n    ## Anchors (positions)\n\n    - `^` – Start of string/line.  \n      - Example: `^Hello` matches “Hello world” but not “He said Hello”.[7][3]\n    - `$` – End of string/line.  \n      - Example: `world$` matches “Hello world” but not “world says hi”.[7][3]\n    - `\\b` – Word boundary.  \n      - Example: `\\bcat\\b` matches “cat” but not “scatter”.[3]\n\n    ## Groups and alternation\n\n    - `(abc)` – Capturing group, treats `abc` as a unit and stores the match.  \n      - Example: `(ab)+` matches “ab”, “abab”, “ababab”.[4][3]\n    - `(?:abc)` – Non‑capturing group (groups for logic, not for capture).  \n      - Example: `(?:ab)+c` matches “abc”, “ababc”.[3]\n    - `a|b` – Alternation, matches `a` or `b`.  \n      - Example: `cat|dog` matches “cat” or “dog”. [8][5]  \n\n    ## Dot and escaping\n\n    - `.` – Any single character except newline (engine‑dependent).  \n      - Example: `a.c` matches “abc”, “a-c”, “a c”.[8][5]\n    - `\\` – Escape to treat a special char as literal.  \n      - Example: `\\.` matches a literal dot in “example.com”.[9][3]\n\n    ## Lookaround (advanced but useful)\n\n    - `(?=...)` – Positive lookahead: must be followed by `...`, but does not consume it.  \n      - Example: `\\d+(?=px)` matches “12” in “12px”.[3]\n    - `(?!...)` – Negative lookahead: must not be followed by `...`.  \n      - Example: `\\d+(?!px)` matches “12” in “12kg” but not in “12px”.[3]\n    - `(?<=...)` / `(?<!...)` – Positive/negative lookbehind (similar but “before” the match).  \n      - Example: `(?<=\\$)\\d+` matches “100” in “$100”.[3]\n\n    ## Tiny end‑to‑end examples\n\n    - Email-ish: `\\w+@\\w+\\.\\w+` matches basic emails like “user@mail.com”.[10]\n    - Simple IPv4: `\\d{1,3}(\\.\\d{1,3}){3}` matches “192.168.0.1”.[11][5]'
